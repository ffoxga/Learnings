### 1. Coin Flips

```python
# head=1, tail=0
np.random.randint(low=0, high=2, size=10) # tossing coin 10 times
```

```python
experiment = np.random.randint(low=0, high=2, size=10)
print(experiment.sum(), 'heads')
```

```python
coin_matrix = np.random.randint(0, 2, size=(10000, 10))
coin_matrix[:5, :]
```

```python
counts = coin_matrix.sum(axis=1)
print(counts[:25])
print(counts.mean())
print(np.median(counts))
print(counts.min(), counts.max())
print(counts.std())
```

```python
np.bincount(counts)
```

```python
unique_numbers = np.arange(11)
observed_times = np.bincount(counts)
print('='*40+'\n')
for n, count in zip(unique_numbers, observed_times):
    print(f'{n} heads observed {count} times {100*count/2000:0.1f}%')
print('\n'+'='*40)
```

### 2. Simulating Stock Returns

```python
import matplotlib.pyplot as plt
```

```python
returns = np.random.normal(0.001, 0.02, 250)
returns[:20]
```

```python
initial_price = 100
```

```python
price = initial_price * np.exp(returns.cumsum())
```

```python
plt.plot(price)
plt.grid();
```

```python
returns = np.random.normal(0.001, 0.02, 250)
initial_price=100
price = initial_price*np.exp(returns.cumsum())
plt.plot(price)
plt.grid();
```
